```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Travel details dataset.csv')
df.info()
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
df.head()
```

```python
df['Duration (days)'] = df['Duration (days)'].astype(int)
df['Start date'] = pd.to_datetime(df['Start date'])
df['End date'] = pd.to_datetime(df['End date'])


df['Accommodation cost'] = df['Accommodation cost'].astype(str).str.replace('$', '', regex=False)
df['Accommodation cost'] = df['Accommodation cost'].str.replace(',', '', regex=False)
df['Accommodation cost'] = df['Accommodation cost'].str.replace('USD', '', regex=False)
df['Accommodation cost'] = df['Accommodation cost'].astype(int)


df['Transportation cost'] = df['Transportation cost'].astype(str).str.replace('$', '', regex=False)
df['Transportation cost'] = df['Transportation cost'].str.replace(',', '', regex=False)
df['Transportation cost'] = df['Transportation cost'].str.replace('USD', '', regex=False)
df['Transportation cost'] = df['Transportation cost'].astype(int)
```

```python
df.info()
```

```python
df.head(5)
```

```python
import re

# cari nama kolom destination (case-insensitive)
dest_col = next((c for c in df.columns if c.lower() == 'destination'), None)
if dest_col is None:
    raise KeyError("Kolom 'destination' tidak ditemukan. Periksa nama kolom di df.columns")

def normalize_destination(x):
    if pd.isna(x):
        return x
    s = str(x).strip()
    # samakan pemisah menjadi koma
    s = re.sub(r'[\-–—/|]+', ',', s)
    # ambil bagian sebelum koma (biasanya nama kota)
    parts = [p.strip() for p in s.split(',') if p.strip()]
    main = parts[0] if parts else s
    # hapus kata negara umum (case-insensitive)
    main = re.sub(r'\b(indonesia|republik indonesia|usa|united states|united kingdom|uk)\b', '', main, flags=re.I).strip()
    # normalisasi spacing dan kapitalisasi nama tempat
    return ' '.join(main.split()).title()

# terapkan normalisasi
df[dest_col] = df[dest_col].apply(normalize_destination)

# lihat hasil setelah normalisasi
print(df[dest_col].value_counts(dropna=False))
```

```python
df.to_csv('Travel details dataset cleaned.csv', index=False)
```
