```python
import pandas as pd
import numpy as np
import seaborn as sns
from jupyter_core.migrate import regex
```

```python
df = pd.read_csv('Travel details dataset.csv')
df.info()
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
df.head()
```

```python
df['Duration (days)'] = df['Duration (days)'].astype(int)
df['Start date'] = pd.to_datetime(df['Start date'])
df['End date'] = pd.to_datetime(df['End date'])


df['Accommodation cost'] = df['Accommodation cost'].astype(str).str.replace('$', '', regex=False)
df['Accommodation cost'] = df['Accommodation cost'].str.replace(',', '', regex=False)
df['Accommodation cost'] = df['Accommodation cost'].str.replace('USD', '', regex=False)
df['Accommodation cost'] = df['Accommodation cost'].astype(int)


df['Transportation cost'] = df['Transportation cost'].astype(str).str.replace('$', '', regex=False)
df['Transportation cost'] = df['Transportation cost'].str.replace(',', '', regex=False)
df['Transportation cost'] = df['Transportation cost'].str.replace('USD', '', regex=False)
df['Transportation cost'] = df['Transportation cost'].astype(int)
```
